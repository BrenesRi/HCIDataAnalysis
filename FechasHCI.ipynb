#**Fechas de Proyectos DCEM**

```python
# @title
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Filtrar las ubicaciones donde 'F323' es 'No Aplica' (tanto en mayúsculas como minúsculas)
df_filtrado = df[df['F323'].str.lower() == 'no aplica']  # .str.lower() asegura que se ignore el caso

# Verifica que la columna 'Ubicacion' esté presente
ubicaciones = df_filtrado['Ubicacion'].unique()

# Si no hay ubicaciones, muestra un mensaje para diagnosticar
if len(ubicaciones) == 0:
    print("No hay ubicaciones con 'No Aplica' en la columna F323")
else:
    dropdown = widgets.Dropdown(
        options=ubicaciones,
        description='Ubicación:',
        disabled=False,
    )

# Función para mostrar la información de las columnas seleccionadas
def mostrar_info(ubicacion):
    # Filtrar el DataFrame para obtener la fila de la ubicación seleccionada
    fila = df[df['Ubicacion'] == ubicacion]

    # Seleccionar las columnas que quieres mostrar
    columnas_seleccionadas = ['FInicioIF313', 'FInicioIF314', 'FInicioIF3225', 'FInicioIF324', 'FInicioIF4', 'FInicioIF5', 'FchInag']

    # Renombrar las columnas con los nuevos nombres
    nuevos_nombres = {
        'FInicioIF313': 'Orden de inicio (Anteproyecto)',
        'FInicioIF314': 'Anteproyecto',
        'FInicioIF3225': 'Orden de inicio (Diseño)',
        'FInicioIF324': 'Diseño',
        'FInicioIF4': 'Construcción',
        'FInicioIF5': 'Puesta en Marcha',
        'FchInag': 'Entrega'
    }

    # Renombrar las columnas de la selección
    info_seleccionada = fila[columnas_seleccionadas].rename(columns=nuevos_nombres)

    # Mostrar la información con los nuevos nombres
    display(info_seleccionada)

# Vincular la función al cambio de selección en el desplegable
widgets.interactive(mostrar_info, ubicacion=dropdown)
```

#**Filtrado General por fechas**

```python
# @title
# Seleccionar las columnas que quieres mostrar
columnas_seleccionadas = ['Ubicacion', 'FInicioIF313', 'FInicioIF314', 'FInicioIF3225', 'FInicioIF324', 'FInicioIF4', 'FInicioIF5', 'FchInag']

# Renombrar las columnas con los nuevos nombres
nuevos_nombres = {
    'FInicioIF313': 'Orden de inicio (Anteproyecto)',
    'FInicioIF314': 'Anteproyecto',
    'FInicioIF3225': 'Orden de inicio (Diseño)',
    'FInicioIF324': 'Diseño',
    'FInicioIF4': 'Construcción',
    'FInicioIF5': 'Puesta en Marcha',
    'FchInag': 'Entrega'
}

# Seleccionar las columnas del DataFrame y renombrarlas
df_seleccionado = df[columnas_seleccionadas].rename(columns=nuevos_nombres)

# Mostrar todas las filas con las columnas renombradas
df_seleccionado
```

```python
# @title
import pandas as pd
import altair as alt

# Convert date columns to datetime objects
date_columns = ['Orden de inicio (Anteproyecto)', 'Anteproyecto', 'Orden de inicio (Diseño)', 'Diseño', 'Construcción', 'Puesta en Marcha', 'Entrega']
# Convert date columns to datetime objects with a specific format
for col in date_columns:
    df_seleccionado[col] = pd.to_datetime(df_seleccionado[col], errors='coerce', format='%d/%m/%Y %H:%M:%S')



# Create a long-form dataframe for plotting
df_long = pd.melt(df_seleccionado, id_vars='Ubicacion', value_vars=date_columns, var_name='Etapa', value_name='Fecha')


# Create a chart
chart = alt.Chart(df_long).mark_point().encode(
    x='Fecha',
    y='Ubicacion',
    color='Etapa'
).properties(
    title='Linea de tiempo de etapas de proyecto por Ubicación',
    width=600,
    height=400
)

chart
```
